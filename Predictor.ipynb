```python
import pandas as pd

# Load CSV fresh
car_df = pd.read_csv("quikr_car.csv")

# Filter valid years
car_df = car_df[car_df['year'].astype(str).str.fullmatch(r'\d{4}')]
car_df['year'] = car_df['year'].astype(int)

# Clean Price safely:
# Extract digits using extract()[0] to get Series
price_cleaned = car_df['Price'].str.replace(',', '').str.extract(r'(\d+)')[0]

# Step 2: Keep only rows with valid price
car_df = car_df[price_cleaned.notnull()].copy()

# Step 3: Assign cleaned and converted price
car_df['Price'] = price_cleaned[price_cleaned.notnull()].astype(int)
```

```python
car_df.info()
```

```python
# Step 1: Remove commas and extract digits
kms_cleaned = car_df['kms_driven'].str.replace(',', '').str.extract(r'(\d+)')[0]

# Step 2: Drop rows with NaN values in kms
car_df = car_df[kms_cleaned.notnull()].copy()

# Step 3: Assign cleaned column and convert to int64
car_df['kms_driven'] = kms_cleaned[kms_cleaned.notnull()].astype('int64')
```
